```python
# the how in both merge, join and concat works like set inner join=interception i.e joining based on like terms, outer means joining everything, left means joining the data set in the sense that the data of the first data set all appears and that on the second data set appears only based on the like ones to the first data set, Right join works like Left but applies only to the second data set
```

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\somte\Desktop\alex the analyst\pandas\LOTR.csv")
```

```python
df2= pd.read_csv(r"C:\Users\somte\Desktop\alex the analyst\pandas\LOTR 2.csv")
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how = 'inner')
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID'])
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'outer', on = ['FellowshipID'])
```

```python
df1.merge(df2, how = 'outer', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'cross')
```

```python
df1.join(df2, how = 'outer', on = 'FellowshipID', lsuffix= '_x', rsuffix='_y')
```

```python
df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), on = 'FellowshipID', lsuffix= '_x', rsuffix='_y')
```
