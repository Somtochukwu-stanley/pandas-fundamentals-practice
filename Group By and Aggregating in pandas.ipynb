```python
#Group by allows you to group together all the values in a column and displays them on the same row, and this is going to allow you to perform aggregate functions on those grouping
```

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\somte\Desktop\alex the analyst\pandas\Flavors.csv")
```

```python
df
```

```python
mean_group_by_frame = df.groupby('Base Flavor')
```

```python
mean_group_by_frame.mean(numeric_only = True)
```

```python
 df.groupby('Base Flavor').mean(numeric_only = True)
```

```python
max_group_by = df.groupby('Base Flavor').max() #it just selects the values that are the highest from the two groups, both strings and integers. it does not pick correspondingly but on group, then individual basis
```

```python
max_group_by 
```

```python
min_group_by = df.groupby('Base Flavor').min()  #it just selects the values that are the lowest from the two groups, both strings and integers. it does not pick correspondingly but on group, then individual basis
```

```python
min_group_by
```

```python
count_group_by =df.groupby('Base Flavor').count()
```

```python
count_group_by
```

```python
sum1_group_by =df.groupby('Base Flavor').sum(['flavor Rating', 'Texture Rating', 'Total Rating'])
```

```python
sum1_group_by
```

```python
agg1_group_by = df.groupby('Base Flavor').agg({'Flavor Rating':['mean', 'max', 'min', 'count', 'sum'], 'Texture Rating':['mean', 'max', 'min', 'count', 'sum']}) #working on just one column
```

```python
agg1_group_by
```
