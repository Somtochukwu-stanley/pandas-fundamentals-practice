```python
df["Phone_Number"] = df["Phone_Number"].astype(str)
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("-", " ", regex = False)
```

```python
def standardize_nigerian_number(num):
    if not num or num.lower() in ["none","n\a", "invalid", " "]:
        return np.nan
    if num.startswith ("0"):
        return "+234" + num [1:]
    if num.startswith ("+234"):
        return num
    if num.startswith ("234"):
        return "+" + num
    return np.nan

df["Phone_Number"] = df["Phone_Number"].apply(standardize_nigerian_number)
```
