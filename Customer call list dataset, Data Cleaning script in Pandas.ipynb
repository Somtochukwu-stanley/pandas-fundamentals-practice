```python
#df["Last_Name"] = df["Last_Name"].str.lstrip("...").. #here the l and r attached to the strip means left and right strip, meaning it should strip off 
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")... #unwanted attachments on the left and right sides of last_name, so you l and r depending on where 
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")... #you want it to function. lstrip is basically for removing unwanted things from our wordings

df["Last_Name"] = df["Last_Name"].str.strip("123._/") #best way to do it
```

```python
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex = True) #that Capital A small a and CO. means that it is telling pandas to remove anything not alphabet and numbers. regex = True confirms it
```

```python
df
```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: str(x))
```

```python
df
```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"]=df["Phone_Number"].str.replace('Na--','')
```

```python
df
```

```python
df["Address"].str.split(',',expand=True) #here it splits the words in a column that has words separated by a comma
```

```python
df[["Street Address","State","Zipcode"]]=df["Address"].str.split(',',expand=True) #it gave the new columns formed above a name
```

```python
df
```

```python
df["Paying Customer"]=df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"]=df["Paying Customer"].str.replace('No','N')
df
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
df = df.replace('N/a','') #it was used because of the backlash
df = df.fillna('')  #it was used because Pandas itself did set the NaN
```

```python
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace = True)
        
df       
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace = True)
df     
#another way to drop null rows
#df = df.dropna(subset = "Phone_Number", inplace =True) 
```
